```python
import json,os
from collections import defaultdict
from tqdm import tqdm
```

```python
def read_kg(kg_path,kg):
    #'/home/xhsun/NLP/KGQA/KG/kgCLUE/Knowledge.txt'
    with open(kg_path) as f:
        lines=f.readlines()

    print('The number of triples: {}'.format(len(lines)))

    sub_map = defaultdict(set)#每一个头实体作为key，对应的所有一跳路径内的(关系，尾实体)作为value
    alias_map=defaultdict(set)
    #ent_to_relations=defaultdict(set)
    bad_line=0
    spliter='|||' if kg=='nlpcc' else '\t'
    for i in tqdm(range(len(lines))):
        line=lines[i]
        l = line.strip().split(spliter)
        s = l[0].strip()
        p = l[1].strip()
        o = l[2].strip()
        if s=='' or p=='' or o=='':
            bad_line+=1
            continue
        sub_map[s].add((p, o))

        #ent_to_relations[s].add(p)

        entity_mention=s
        if kg.lower()=='kgclue' and ('（' in s and '）' in s):
            entity_mention=s.split('（')[0]
            alias_map[entity_mention].add(s)
        if kg.lower()=='nlpcc' and ('(' in s and ')' in s):
            entity_mention=s.split('(')[0]
            alias_map[entity_mention].add(s)

        if p in ['别名','中文名','英文名','昵称','中文名称','英文名称','别称','全称','原名']:
            alias_map[entity_mention].add(o)
    return alias_map,sub_map
```

```python
kg_path='/home/xhsun/NLP/KGQA/KG/kgCLUE/Knowledge.txt'
kg='kgclue'
alias_map,sub_map=read_kg(kg_path,kg)
```

# 处理NER数据

```python
def process_ner_data(data,ner_data,questions_set):
    for example in data:
        context,qas=example['context'],example['qas']
        if qas==[]:
            continue
        for qa in qas:
            entity=qa['answers'][0]['text']
            if entity in alias_map or entity in sub_map:
                if context not in questions_set:
                    questions_set.append(context)
                    text='{}'.format(context)+'。上个句子中哪些属于实体名词？_。'
                    ner_data.append([text,entity])

path_folder='/home/xhsun/NLP/NLPdata/MRC/squad2.0-format/all_train_files/ner/'
ner_data=[]
questions_set=[]
for file_name in os.listdir(path_folder):
    file=os.path.join(path_folder,file_name)
    try:
        with open(file) as f:
            data=json.load(f)['data'][0]['paragraphs']
            print("{} has {} examples".format(file,len(data)))
            process_ner_data(data,ner_data,questions_set)
    except:
        print(file,'-'*100)
        continue
```

# 处理MRC数据

```python
def process_mrc_data(data,mrc_data):
    for example in data:
        context,qas=example['context'],example['qas']
        if qas==[]:
            continue
        for qa in qas:
            answer=qa['answers'][0]['text']
            question=qa['question']
            text="{}。上个句子的答案是？_。".format(question)
            mrc_data.append([text,answer])

path_folder='/home/xhsun/NLP/NLPdata/MRC/squad2.0-format/all_train_files/mrc/'
mrc_data=[]
for file_name in os.listdir(path_folder):
    file=os.path.join(path_folder,file_name)
    #try:
    with open(file) as f:
        data=json.load(f)['data'][0]['paragraphs']
        print("{} has {} examples".format(file,len(data)))
        process_mrc_data(data,mrc_data)
#     except:
#         print(file,'-'*100)
#         continue
```

## 处理WebQA数据集

```python
webqa_path='/home/xhsun/NLP/NLPdata/MRC/WebQA/WebQA.v1.0'

def process_webqa_data(file_path):
    webqa_data=[]
    tmp_data=set()
    with open(file_path) as f:
        data=json.load(f)

        for example in list(data.values()):
            question=example['question']
            evidences=list(example['evidences'].values())
            for answer_evidence in evidences:
                answer,evidence=answer_evidence['answer'],answer_evidence['evidence']
                if answer==['no_answer']:
                    continue
                else:
                    answer=answer[0]
                    tmp_data.add((question,answer))
    for q,a in tmp_data:
        webqa_data.append(["{}。上个句子的答案是？_。".format(q),a])
    return webqa_data

webqa_data=[]
webqa_data+=process_webqa_data(file_path='/home/xhsun/NLP/NLPdata/MRC/WebQA/WebQA.v1.0/me_train.json')
webqa_data+=process_webqa_data(file_path='/home/xhsun/NLP/NLPdata/MRC/WebQA/WebQA.v1.0/me_validation.ir.json')
webqa_data+=process_webqa_data(file_path='/home/xhsun/NLP/NLPdata/MRC/WebQA/WebQA.v1.0/me_validation.ann.json')
```
