```python
examples=[]
with open('/home/xhsun/Desktop/graduate_models/Section4/kgclue/mixture-predictions.json') as f:
    lines=f.readlines()
for line in lines:
    examples.append(json.loads(line.strip()))
```

```python
len(examples)
```

```python
examples[0]
```

```python
qa_examples=[]
ner_examples=[]
for example in examples:
    question=example['question']
    if '上个句子中哪些属于实体名词？' in question:
        ner_examples.append(example)
    else:
        qa_examples.append(example)
```

```python
print(len(ner_examples),len(qa_examples))
```

```python
uncorrect_examples_qa=[]
TP=0
FN=0
FP=0
FP_examples=[]
FN_examples=[]
for example in qa_examples:
    answer=example['answer'][0].lower()
    predict=example['predict'].lower()
    question=example['question'].lower()
    q,choice=question.split("这个句子的意图与下列哪一个关系最相似？_。")
    if answer==predict:
        TP+=1
    else:
        if answer=='不匹配':
            #将负样本预测为正样本,特别注意，数据集存在问题
            if predict not in choice:
                FP+=1
                FP_examples.append(example)
        else:
            assert answer!='不匹配'
            #将正样本识别错误
            FN+=1
            FN_examples.append(example)
        uncorrect_examples_qa.append(example)
```

```python
print(len(uncorrect_examples_qa),TP,FP,FN,TP/(TP+FN),TP/(TP+FP))
```
