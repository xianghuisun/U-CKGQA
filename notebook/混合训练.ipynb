```python
import pandas as pd
import json
import random
```

```python
qa_train=pd.read_csv('/data/aisearch/nlp/data/xhsun/seq2seqUnify/data/KGQA/QA/prompt_train.csv')
qa_test=pd.read_csv('/data/aisearch/nlp/data/xhsun/seq2seqUnify/data/KGQA/QA/prompt_test.csv')

ner_train=pd.read_csv('/data/aisearch/nlp/data/xhsun/seq2seqUnify/data/KGQA/NER/prompt_train.csv')
ner_test=pd.read_csv('/data/aisearch/nlp/data/xhsun/seq2seqUnify/data/KGQA/NER/prompt_test.csv')
```

# 混合训练

```python
print(qa_test.values.shape,qa_train.values.shape,ner_test.values.shape,ner_train.values.shape)
```

```python
kgqa_train=qa_train.values.tolist()+ner_train.values.tolist()
kgqa_test=qa_test.values.tolist()+ner_test.values.tolist()
random.shuffle(kgqa_train)
print(len(kgqa_train),len(kgqa_test))
```

```python
kgqa_train[:5]
```

```python
columns=['question','label']
kgqa_train=pd.DataFrame(kgqa_train,columns=columns)
kgqa_train.to_csv('/data/aisearch/nlp/data/xhsun/seq2seqUnify/data/KGQA/mix_train.csv',index=None)
kgqa_test=pd.DataFrame(kgqa_test,columns=columns)
kgqa_test.to_csv('/data/aisearch/nlp/data/xhsun/seq2seqUnify/data/KGQA/mix_test.csv',index=None)
```

# 分析测试结果(test_public)

```python
with open('../../saved_models/kgqa/mixture/mixture-predictions.json') as f:
    lines=f.readlines()
test_public_examples=[]
for line in lines:
    test_public_examples.append(json.loads(line.strip()))
```

```python
wrong_examples=[]
for example in test_public_examples:
    predict=example['predict']#.lower()
    answer=example['answer'][0]#.lower()
    if predict!=answer:
        wrong_examples.append(example)
```
